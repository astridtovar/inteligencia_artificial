```python
# IMPORTAMOS LAS LIBRERIAS QUE VAMOS A USAR
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
```

```python
# Usamos la función read_csv de pandas para cargar la base de datos
url = 'https://raw.githubusercontent.com/astridtovar/inteligencia_artificial/main/anemia.csv'
bd = pd.read_csv(url)

# Mostramos la bd
bd.head()
```

```python
# Mostramos algunas gráficas de los datos
sns.pairplot(bd.iloc[:,:],hue='Result')
plt.show()
```

```python
# Dividimos los datos en variables independientes X, y variable dependiente y (especies)
X = bd.drop(["Result"],axis = 1) #axis refiere a columna
y = bd.Result
```

```python
# Dividimos datos para entrenamiento y prueba

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # test_size define el conjunto de prueba como un 30%
```

```python
# Especificamos nuestro modelo
from sklearn.linear_model import LogisticRegression

logisticRegr = LogisticRegression(max_iter = 1000) # iteraciones del algoritmo de optimización
```

```python
#Entrenamos el modelo
logisticRegr.fit(X_train, y_train)
```

```python
# Predicción del modelo

from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix

ypred = logisticRegr.predict(X_test)
matriz = confusion_matrix(y_test,ypred)

plot_confusion_matrix(conf_mat=matriz, figsize=(4,4), show_normed=False)
plt.tight_layout()
```
